```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = sns.load_dataset('titanic')
df.head()
```

```python
fig1 = sns.scatterplot(x="age", y="fare", hue="survived", data=df)
plt.title('Scatter Plot of Titanic Dataset')
# save this plot as png file in the Plots folder
plt.savefig("./Plots/fig1.png")
```

```python
fig2 = sns.lineplot(x="age", y="fare", data=df)
plt.title('Line Plot of Titanic Dataset')
# save this plot as png file in the Plots folder
plt.savefig("./Plots/fig2.png")
```

```python
fig3 = sns.barplot(x="who", y="fare", hue='class', data=df)
plt.title('Bar Plot of Titanic Dataset')
# save this plot as png file in the Plots folder
plt.savefig("./Plots/fig3.png")
```

```python
fig4 = sns.histplot(data=df, x='age', multiple='stack', kde=True)
plt.title('Histogram of Titanic Dataset')
# save this plot as png file in the Plots folder
plt.savefig("./Plots/fig4.png")
```

```python
sns.boxplot(df, x='age', hue='sex')
plt.title('Box Plot of Titanic Dataset')
# save this plot as png file in the Plots folder
plt.savefig("./Plots/fig4.png")
```

```python
# Exclude non-numeric columns from the DataFrame
numeric_df = df.select_dtypes(include=[float, int])

# Create the heatmap using the numeric DataFrame
sns.heatmap(numeric_df)
plt.title('Heatmap of Titanic Dataset')
plt.savefig("./Plots/fig5.png")
```

```python
sns.pairplot(df)
plt.title('Pair Plot of Titanic Dataset')
# save this plot as png file in the Plots folder
plt.savefig("./Plots/fig6.png")
```

```python
sns.relplot(x='age', y='fare', data=df, hue='class', col='sex', col_wrap=2, kind='scatter')
```
